```python
rf_y_pred = rf.predict(X_test)
```

```python
#Accuracy testing
predicted_mean = minmax_scale(rf_y_pred,feature_range=(0,1))
y_test_mean = minmax_scale(y_test,feature_range=(0,1))

rf_mse = mean_squared_log_error(y_test_mean,predicted_mean)
rf_mse = np.sqrt(rf_mse)
rf_mse = rf_mse

round(rf_mse*100)
```
